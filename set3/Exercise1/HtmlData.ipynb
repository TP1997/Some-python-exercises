```python
import pandas as pd
from IPython.display import display as disp 

gdpData = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)_per_capita", index_col=1, header=0, decimal=",", thousands='s')
gdpData = gdpData[2]
gdpData.columns=['Rank', 'GDP']

happyData = pd.read_html("https://en.wikipedia.org/wiki/World_Happiness_Report", header=0)
happyData = happyData[4][['Country or region', 'Score']]
happyData.columns=['Country', 'Score']

lifeData = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_life_expectancy", index_col=0, header=0)
lifeData = lifeData[0][['Both sexesrank', 'Both sexes lifeexpectancy']]
lifeData.columns=['Rank', 'Expectancy']

birthData = pd.read_html("https://en.wikipedia.org/wiki/List_of_sovereign_states_and_dependent_territories_by_birth_rate", index_col=0, header=0,skiprows=1)
birthData=pd.DataFrame(birthData)
#birthData = birthData[1][['CIA WF 2016']]
```
